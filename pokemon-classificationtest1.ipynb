```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 설정 (Windows 기준)
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 맑은 고딕
# font_path = 'C:/Windows/Fonts/gulim.ttf'  # 굴림체

# 폰트 설정
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

import pandas as pd
import os
from sklearn.model_selection import train_test_split

# 포켓몬 타입 리스트
pokemon_types = ['Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 
                'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic', 
                'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']

# 기본 설정
image_size = (150, 150)
batch_size = 32
epochs = 20  # 에포크 수 늘려서 더 오래 학습
    
# 데이터 경로 설정
base_dir = ".\pokemon_by_type"  # 데이터 폴더 경로
    
# 데이터 준비
print("데이터 준비 중...")

# 폴더 구조에서 DataFrame 생성 (타입별로 폴더가 있는 경우)
data = []
    
# 각 타입 폴더 검색
for type_name in pokemon_types:
    type_dir = os.path.join(base_dir, type_name.lower())
    if not os.path.exists(type_dir):
        print(f"{type_name} 폴더 없음, 건너뜁니다.")
        continue
        
    # 해당 타입 폴더의 모든 이미지 파일 검색
    for img_file in os.listdir(type_dir):
        if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(type_dir, img_file)
            
            # 새 데이터 행 생성
            row = {'filepath': img_path}
            
            # 모든 타입 초기화
            for t in pokemon_types:
                row[t] = 0
            # 현재 타입만 1로 설정
            row[type_name] = 1
            
            # 다중 타입 처리 (파일명에 타입이 포함되어 있다면)
            # 예: "bulbasaur_grass_poison.jpg"
            file_name = os.path.splitext(img_file)[0].lower()
            for t in pokemon_types:
                if t.lower() in file_name and t.lower() != type_name.lower():
                    row[t] = 1
            
            data.append(row)
    
# DataFrame 생성
df = pd.DataFrame(data)
print(f"총 {len(df)} 개의 이미지 찾음")

# 데이터 분할
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# 데이터 증강 설정
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
)
print(df.columns)

#train_datagen = ImageDataGenerator(rescale=1./255) 
val_test_datagen = ImageDataGenerator(rescale=1./255)

# DataFrame에서 데이터 생성
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    #directory=base_dir if 'filepath' in df.columns else None,
    directory='./' if 'filepath' in df.columns else None,
    x_col='filepath',
    y_col=pokemon_types,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw',  # 다중 레이블 분류
    shuffle=True
)
    
val_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    #directory=base_dir if 'filepath' in df.columns else None,
    directory='./' if 'filepath' in df.columns else None,
    x_col='filepath',
    y_col=pokemon_types,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw',
    shuffle=False
)
    
test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    #directory=base_dir if 'filepath' in df.columns else None,
    directory='./' if 'filepath' in df.columns else None,
    x_col='filepath',
    y_col=pokemon_types,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw',
    shuffle=False
)

print(f"훈련용 데이터 수: {train_generator.samples}")
print(f"검증용 데이터 수: {val_generator.samples}")
print(f"테스트용 데이터 수: {test_generator.samples}")

# 모델 구축
print("모델 구축 중...")
#model = build_simple_cnn_model(input_shape=(*image_size, 3))
# 심플한 CNN 모델 구축 (계속)
input_shape=(150, 150, 3)
num_classes=len(pokemon_types)
model = Sequential([
    Input(shape=(150, 150, 3),),
    # 첫 번째 컨볼루션 블록
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    #Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Dropout(0.2),
     
    # 두 번째 컨볼루션 블록
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    #Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Dropout(0.2),
        
    # 세 번째 컨볼루션 블록
    #Conv2D(128, (3, 3), activation='relu', padding='same'),
    #Conv2D(128, (3, 3), activation='relu', padding='same'),
    #MaxPooling2D((2, 2)),
    #Dropout(0.25),
        
    # 네 번째 컨볼루션 블록 (더 깊게)
    #Conv2D(256, (3, 3), activation='relu', padding='same'),
    #Conv2D(256, (3, 3), activation='relu', padding='same'),
    #MaxPooling2D((2, 2)),
    #Dropout(0.25),
        
    # 분류기
    Flatten(),
    #Dense(512, activation='relu'),
    #Dropout(0.5),
    Dense(256, activation='relu'),
    #Dropout(0.5),
    Dense(num_classes, activation='sigmoid')  # 다중 레이블 분류를 위한 시그모이드
])
    
# 컴파일
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',  # 다중 레이블 분류
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)
    
model.summary()

# 모델 학습
print("모델 학습 시작...")
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=val_generator,
    callbacks=[
        # 조기 종료 (검증 손실이 10 에포크 동안 개선되지 않으면 학습 중단)
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss', patience=10, restore_best_weights=True
        ),
        # 학습률 감소 (검증 손실이 5 에포크 동안 개선되지 않으면 학습률 감소)
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6
        )
    ]
)

# 학습 및 검증 그래프 그리기
plt.figure(figsize=(12, 4))

# 손실 그래프
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
    
# 정확도 그래프
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

    
# 학습 곡선 시각화
# 모델 저장
model.save('pokemon_type_classifier_cnn.keras')
print("모델 저장 완료: pokemon_type_classifier_cnn.keras")
   
# 테스트 세트 평가
print("테스트 세트 평가 중...")
test_loss, test_acc, test_precision, test_recall = model.evaluate(test_generator)
print(f"테스트 손실: {test_loss:.4f}")
print(f"테스트 정확도: {test_acc:.4f}")

# 포켓몬 타입 예측 함수
test_image_path = "test_pokemon.jpg"
    
img = image.load_img(test_image_path, target_size=image_size)
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0
    
# 예측
predictions = model.predict(img_array)[0]
    
# 이미지 표시
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.axis('off')
plt.title('포켓몬 이미지')
plt.show()
    
# 결과 출력
print("\n예측된 포켓몬 타입:")
    
# 확률 높은 순서대로 정렬
sorted_idx = np.argsort(predictions)[::-1]
    
# 상위 타입 또는 임계값 이상 타입 출력
for i in sorted_idx:
    if predictions[i] > 0.3:  # 30% 이상 확률만 표시
        print(f"{pokemon_types[i]}: {predictions[i]*100:.2f}%")
    
# 그래프로 표시
plt.figure(figsize=(10, 6))
plt.barh(y=[pokemon_types[i] for i in sorted_idx[:5]], 
         width=[predictions[i]*100 for i in sorted_idx[:5]])
plt.xlabel('확률 (%)')
plt.title('포켓몬 타입 예측 결과 (상위 5개)')
plt.gca().invert_yaxis()  # 확률 높은 순으로 위에서 아래로
plt.tight_layout()
plt.show()
```
